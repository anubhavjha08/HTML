```python
from astropy.io import fits
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hdu_list  = fits.open('bbso_halph_fl_20240526_173036.fts')
hdu_list.info()
```

```python
image_data = hdu_list[0].data
```

Our data is now stored as a 2D numpy array. But how do we know the dimensions of the image? We can look at the shape of the array.

```python
print(type(image_data))
print(image_data.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LogNorm, Normalize


plt.figure()
vmin = np.percentile(image_data, 1)  # Lower limit for color map
vmax = np.percentile(image_data, 99)  # Upper limit for color map
plt.imshow(image_data, origin='lower', cmap='Greys', vmin=vmin, vmax=vmax)

# plt.imshow(image_data, origin='lower', norm= LogNorm(), cmap = 'Greys')
plt.colorbar()
plt.show()
```

```python
print('Min:', np.min(image_data))
print('Max:', np.max(image_data))
print('Mean:', np.mean(image_data))
print('Stdev:', np.std(image_data))
```

```python
image_data_flat =image_data.flatten()  # cast the data from a 2D array to a 1D numpy array
print(type(image_data.flat))
print(image_data.flatten().shape)
```
